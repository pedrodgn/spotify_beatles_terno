```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')
```

```python
terno_features_df = pd.read_csv("terno_track_features.csv")
beatles_features_df = pd.read_csv("beatles_track_features.csv")
```

```python
beatles_features_df = beatles_features_df[~beatles_features_df["album_name"].str.contains("Live At")]
beatles_features_df = beatles_features_df[~beatles_features_df["album_name"].str.contains("Rooft")]
beatles_features_df = beatles_features_df[~beatles_features_df["album_name"].str.contains("Deluxe")]
beatles_features_df = beatles_features_df[~beatles_features_df["album_name"].str.contains("Songtrack")]
beatles_features_df = beatles_features_df[~beatles_features_df["album_name"].str.contains("Naked")]
beatles_features_df = beatles_features_df[~beatles_features_df["name"].str.contains("Mix")]
beatles_features_df = beatles_features_df[~beatles_features_df["name"].str.contains("Demo")]
```

```python
non_vocal_songs = ['Pepperland - Remastered 2009', 'Sea Of Time - Remastered 2009',
       'Sea Of Holes - Remastered 2009',
       'Sea Of Monsters - Remastered 2009',
       'March Of The Meanies - Remastered 2009',
       'Pepperland Laid Waste - Remastered 2009',
       'Yellow Submarine In Pepperland - Remastered 2009']
```

```python
beatles_features_df = beatles_features_df[~beatles_features_df["name"].isin(non_vocal_songs)]
```

```python
full_features_df = pd.concat([terno_features_df,beatles_features_df]).reset_index(drop=True)
```

```python
features = ['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness',
       'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']
```

```python
x = full_features_df[features]
```

```python
wcss = []
for i in range(1,11):
    model = KMeans(n_clusters = i, init = "k-means++")
    model.fit(x)
    wcss.append(model.inertia_)
plt.figure(figsize=(10,10))
plt.plot(range(1,11), wcss)
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
scaler = StandardScaler()
scaler.fit(x)
x = scaler.transform(x)
```

```python
pca = PCA(3)
data = pca.fit_transform(x)
```

```python
plt.figure(figsize=(10,10))
var = np.round(pca.explained_variance_ratio_*100, decimals = 1)
lbls = [str(x) for x in range(1,len(var)+1)]
plt.bar(x=range(1,len(var)+1), height = var, tick_label = lbls)
plt.show()
```

```python
n_clusters = 4
```

```python
model2 = KMeans(n_clusters = n_clusters, init = "k-means++")
model2.fit(data)
```

```python
import seaborn as sns
```

```python
centers = np.array(model2.cluster_centers_)
model = KMeans(n_clusters = n_clusters, init = "k-means++")
label = model.fit_predict(data)
plt.figure(figsize=(10,10))
uniq = np.unique(label)
for i in uniq:
    plt.scatter(data[label == i , 0] , data[label == i , 1] , label = i)
plt.scatter(centers[:,0], centers[:,1], marker="x", color='k')

#This is done to find the centroid for each clusters.
plt.legend()
plt.show()
```

```python
full_features_df["cluster"] = label
pca_df = pd.DataFrame(data,columns=["PCA_0","PCA_1","PCA_2"])

full_features_df = full_features_df.join(pca_df)
```

```python
from sklearn.metrics.pairwise import euclidean_distances
```

```python
import plotly.express as px

fig = px.scatter(full_features_df,x="PCA_0",y="PCA_1",color="cluster",hover_data=['name', 'album_name', 'artist_name'])
fig.show()
```

```python
def get_2_songs(x,full_features_df):
    
    euclidean_dist = full_features_df.apply(lambda x_1: euclidean_distances([[x["PCA_0"],x["PCA_1"]]],[[x_1["PCA_0"],x_1["PCA_1"]]]),axis=1)
    full_features_df["test_dist"] = euclidean_dist.apply(lambda x: x[0][0])
    
    full_features_df = full_features_df.sort_values("test_dist")
    
    return list(full_features_df[full_features_df["artist_name"]!=x["artist_name"]]["name"][:2])
```

```python
full_features_df["reccomendations"] = full_features_df.apply(lambda x: get_2_songs(x,full_features_df),axis=1)
full_features_df["reccomendations_0"] = full_features_df["reccomendations"].apply(lambda x: x[0])
full_features_df["reccomendations_1"] = full_features_df["reccomendations"].apply(lambda x: x[1])
```
